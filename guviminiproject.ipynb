```python
import mysql.connector

# Connect to MySQL (XAMPP)
conn = mysql.connector.connect(
    host="localhost",
    user="root",  # Default XAMPP username
    password="",  # Default XAMPP has no password
    database="retail_orders"  # Use your actual database name
)

cursor = conn.cursor()

# Run SQL query
query = """
    SELECT product_name, SUM(sales_price) AS total_revenue
    FROM orders
    WHERE sales_price IS NOT NULL AND sales_price > 0
    GROUP BY product_name
    ORDER BY total_revenue DESC
    LIMIT 10;
"""
cursor.execute(query)

# Fetch and print results
results = cursor.fetchall()
if not results:
    print("No valid revenue data found.")
else:
    print("\nTop 10 Highest Revenue-Generating Products:")
    for product, revenue in results:
        print(f"{product}: ${revenue:.2f}")

# Close connection
cursor.close()
conn.close()
```

```python
import mysql.connector

try:
    print("🔹 Connecting to MySQL...")
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="retail_orders"
    )
    cursor = conn.cursor()
    print("✅ Connected to MySQL!")

    # Get columns in 'orders'
    cursor.execute("SHOW COLUMNS FROM orders")
    orders_columns = [col[0] for col in cursor.fetchall()]
    print("📌 Orders columns:", orders_columns)

    # Get columns in 'orders_new_dataset'
    cursor.execute("SHOW COLUMNS FROM orders_new_dataset")
    orders_new_columns = [col[0] for col in cursor.fetchall()]
    print("📌 Orders_new_dataset columns:", orders_new_columns)

    # Identify JOIN column
    join_col = None
    if "product_id" in orders_columns and "product_id" in orders_new_columns:
        join_col = "product_id"

    if not join_col:
        print("❌ No common column found for JOIN. Exiting.")
        conn.close()
        exit()

    # Identify sales column
    sales_col = "sales_price" if "sales_price" in orders_new_columns else None
    if not sales_col:
        print("❌ sales_price column not found in orders_new_dataset. Exiting.")
        conn.close()
        exit()

    print(f"🔹 Using JOIN on `{join_col}`, and summing `{sales_col}`.")

    # Create table
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS highest_sales_product (
            product_id VARCHAR(255), 
            total_sales FLOAT
        )
    """)
    print("✅ Table 'highest_sales_product' created.")

    # Clear previous data
    cursor.execute("DELETE FROM highest_sales_product")

    # Query to find the product with the highest total sales
    query = f"""
        INSERT INTO highest_sales_product (product_id, total_sales)
        SELECT o.{join_col}, SUM(n.{sales_col}) AS total_sales
        FROM orders o
        JOIN orders_new_dataset n ON o.{join_col} = n.{join_col}
        GROUP BY o.{join_col}
        ORDER BY total_sales DESC
        LIMIT 1
    """
    print("🔹 Running query...")
    cursor.execute(query)

    # Commit and close connection
    conn.commit()
    conn.close()
    print("✅ Product with the highest total sales inserted successfully!")

except mysql.connector.Error as err:
    print("❌ MySQL Error:", err)

except Exception as e:
    print("❌ Unexpected Error:", e)
```

```python
import mysql.connector

# ✅ Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="retail_orders"
)
cursor = conn.cursor()

# ✅ Check available columns in 'orders_new_dataset'
cursor.execute("SHOW COLUMNS FROM orders_new_dataset")
orders_new_columns = [col[0] for col in cursor.fetchall()]
print("📌 Available columns in 'orders_new_dataset':", orders_new_columns)

# ✅ Ensure 'product_id' and 'sales_price' exist
product_col = "product_id" if "product_id" in orders_new_columns else None
sales_col = "sales_price" if "sales_price" in orders_new_columns else None

if not product_col or not sales_col:
    print("❌ Required columns not found! Check database schema.")
    conn.close()
    exit()

# ✅ SQL Query to Calculate Total Sales Per Product (Without JOIN)
query = f"""
    SELECT {product_col}, SUM({sales_col}) AS total_sales
    FROM orders_new_dataset
    GROUP BY {product_col}
    ORDER BY total_sales DESC;
"""

cursor.execute(query)
result = cursor.fetchall()

# ✅ Print Results
print("\n✅ Total Sales Per Product:")
for row in result:
    print(f"Product ID: {row[0]}, Total Sales: {row[1]:.2f}")

# ✅ Close Connection
conn.close()
```

```python
###group by operations :

import mysql.connector

# ✅ Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",  # No password used
    database="retail_orders"
)
cursor = conn.cursor()

# ✅ Create a new table to store query results
cursor.execute("DROP TABLE IF EXISTS product_sales")
cursor.execute("""
    CREATE TABLE product_sales (
        product_id INT,
        total_sales FLOAT
    )
""")

# ✅ Run SQL Query & Insert Results into the New Table
query = """
    INSERT INTO product_sales (product_id, total_sales)
    SELECT product_id, SUM(sales_price) 
    FROM orders_new_dataset
    GROUP BY product_id
"""
cursor.execute(query)
conn.commit()

print("✅ Query executed successfully, and results stored in 'product_sales' table")

# ✅ Close Connection
cursor.close()
conn.close()
```
